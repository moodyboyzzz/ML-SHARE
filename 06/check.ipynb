```python
from sympy import symbols, diff, solve, simplify
import pandas as pd

# Определение символов
l = symbols('l1:4')
q = symbols('q')
x = [4, 9, 7]
y = [1, 1, -1]

# Определение формулы
formula = 1/2 * sum(l[i]*y[i]*sum(l[j]*y[j]*(x[i]*x[j]+1)**2 for j in range(3)) for i in range(3)) - sum(l[i] for i in range(3))

# Вывод исходной формулы
print("Исходная формула:")
print(formula)

formula_with_substitution = formula.subs('l3', 'l2 + l1')

print("\nФормула с заменой l3 на l2 + l1:")
print(formula_with_substitution)

derivative_l1 = diff(formula_with_substitution, l[0])
derivative_l2 = diff(formula_with_substitution, l[1])

print("Производная по l1:")
print(derivative_l1)
print("\nПроизводная по l2:")
print(derivative_l2)

solution = solve([derivative_l1, derivative_l2], (l[0], l[1]))

print("\nРешение системы уравнений:")
print(solution)
print(solution[l[0]])
l1_value = solution[l[0]]
l2_value = solution[l[1]]
l3_value = l1_value + l2_value

print("Значение l1:", l1_value)
print("Значение l2:", l2_value)
print("Значение l3:", l3_value)

#Формула для w0
formula2 = sum(l[i]*y[i]*(x[i]*x[0]+1)**2 for i in range(3)) - 1
print("\nФормула для w0")
print(formula2)

w0 = formula2.subs({l[0]: l1_value, l[1]: l2_value, l[2]: l3_value})
print("\nЗначение w0:")
print(w0)

formula3 = sum(l[i]*y[i]*(x[i]*q + 1)**2 for i in range(3)) - w0

# Вывод третьей формулы
print("\nТретья формула:")
print(formula3)

formula3_substituted_with_values = formula3.subs({l[0]: l1_value, l[1]: l2_value, l[2]: l3_value})

# Вывод третьей формулы с подставленными значениями переменных
print("\nКонечное выражение:")
print(formula3_substituted_with_values)

simplified_formula3 = simplify(formula3_substituted_with_values)

# Вывод упрощенной формулы
print("\nУпрощенная третья формула:")
print(simplified_formula3)

coefficients_dict = simplified_formula3.as_coefficients_dict()

# Получение коэффициентов
coefficient_q2 = coefficients_dict[q**2] if q**2 in coefficients_dict else 0
coefficient_q = coefficients_dict[q] if q in coefficients_dict else 0
constant_term = coefficients_dict[1] if 1 in coefficients_dict else 0

# Вывод коэффициентов
print("\nКоэффициент при q^2:", coefficient_q2)
print("Коэффициент при q:", coefficient_q)
print("Свободный член:", constant_term)

coefficients = pd.DataFrame([[coefficient_q2, coefficient_q, constant_term]], columns=['c2', 'c1', 'c0'])
coefficients.to_csv('seminar06_svm.csv', index=False)
file = pd.read_csv('seminar06_svm.csv')
print("\n")
print(file)
```
